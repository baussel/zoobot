```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
def get_damage(sig, vel, base_dam, exp_rad, exp_vel, dam_red):
    direct_damage = sig/exp_rad
    bounce_damage = (sig*exp_vel/(exp_rad*vel)) ** dam_red
    return base_dam * np.min([np.ones_like(direct_damage), direct_damage, bounce_damage], axis=0)
```

```python
sig = np.linspace(50, 1000)
vel = np.linspace(0.1, 583)
xx, yy = np.meshgrid(sig, vel)

# sig = np.linspace(250, 5000)
# vel = np.linspace(0.1, 4667)
# xx, yy = np.meshgrid(sig, vel)

base_dam_storm = 167
exp_rad_storm = 107
exp_vel_storm = 158
dam_red_storm = 0.65

base_dam_thun= 346
exp_rad_thun = 250
exp_vel_thun = 118
dam_red_thun = 0.82
```

```python

damage_storm = get_damage(xx, yy, base_dam_storm, exp_rad_storm, exp_vel_storm, dam_red_storm)

damage_thun = get_damage(xx, yy, base_dam_thun, exp_rad_thun, exp_vel_thun, dam_red_thun)
```

```python
fig, ax = plt.subplots()
cf = ax.contourf(xx, yy, damage_storm)
fig.colorbar(cf)
ax.set_xscale('log')
ax.set_xlabel('Sig Radius (m)')
ax.set_ylabel('Target Velocity (m/s)')
ax.set_title('Stormbringer DPS')
```

```python
fig, ax = plt.subplots()
cf = ax.contourf(xx, yy, damage_thun)
fig.colorbar(cf)
ax.set_xscale('log')
ax.set_xlabel('Sig Radius (m)')
ax.set_ylabel('Target Velocity (m/s)')
ax.set_title('Thunderchild DPS')
```
