# what ratios do we get for smooth vs featured?

```python
q1_df = df.query('question == "smooth-or-featured"')
q1_df = q1_df.query('user_id != ""')  # exclude not-logged-in for now
q1_df['response'].value_counts() / len(q1_df)
```

```python
# naming the functions to get the column name

def smooth_fraction(x):
    return np.mean(x == 'smooth')
    
def featured_fraction(x):
    return np.mean(x == 'featured-or-disk')
    
def artifact_fraction(x):
    return np.mean(x == 'artifact')
    
        
```

```python
artifact_ratios = q1_df.groupby('user_id').agg({'response': [smooth_fraction, featured_fraction, artifact_fraction], 'classification_id': 'count'}).reset_index()
```

# Calculate P-values

```python
from scipy.stats import binom
```

```python
def get_pvalue(df, frac_to_use=.2):
    # responses['not_artifact_fraction'] = 1-responses['artifact_fraction']

    # frac_to_use = overall_artifact_frac
    if 'artifact_count' not in df.columns.values:
        df['artifact_count'] = df['artifact_fraction'] * df['count']
#     df['not_artifact_count'] = df['count'] - df['artifact_count']
    df['pvalue'] = 1 - binom.cdf(df['artifact_count']-1, df['count'], frac_to_use)
    return df
```
