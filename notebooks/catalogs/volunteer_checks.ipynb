```python
def get_iauname(metadata):
    possible_keys = ['iauname', '!iauname', 'IAUNAME', '!IAUNAME']
    for key in possible_keys:
        if key in metadata.keys():
            return metadata[key]
    return np.nan
```

```python
def extract_smooth_answer(annotations):
    annotations = json.loads(annotations)
    if annotations[0]['task_label'] == 'Is the galaxy simply smooth and rounded, with no sign of a disk?':
        return annotations[0]['value']
    return np.nan
```

```python
def get_computer_id(x):
    if not pd.isna(x['user_id']): 
        return str(int(x['user_id']))
    if not pd.isna(x['user_ip']):
        return 'ip_' + x['user_ip']
    else:
        return ""
```

```python
def load_df(df_loc, nrows=None):
    df = pd.read_csv(df_loc, nrows=nrows, usecols=['classification_id', 'user_id', 'user_ip', 'created_at', 'annotations', 'subject_data'])
    df = df[pd.to_datetime(df['created_at']) >= datetime(year=2018, month=3, day=15, tzinfo=timezone.utc)]
    df['iauname'] = df['subject_data'].apply(lambda x: get_iauname(list(json.loads(x).values())[0]))
    df['smooth-or-featured_value'] = df['annotations'].apply(extract_smooth_answer)
    df['computer_id'] = df.apply(get_computer_id, axis=1)
#     del df['subject_data']
    del df['annotations']
    return df
```

# Calculate P-values

```python
from scipy.stats import binom
```

```python
def get_pvalue(df, frac_to_use=.2):
    # responses['not_artifact_fraction'] = 1-responses['artifact_fraction']

    # frac_to_use = overall_artifact_frac
    if 'artifact_count' not in df.columns.values:
        df['artifact_count'] = df['artifact_fraction'] * df['count']
#     df['not_artifact_count'] = df['count'] - df['artifact_count']
    df['pvalue'] = 1 - binom.cdf(df['artifact_count']-1, df['count'], frac_to_use)
    return df
```
