```python
import os

import seaborn as sns
sns.set_context('notebook')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Table
```

```python
catalog_loc = 'gz_decals_auto_posteriors.parquet'
ml = pd.read_parquet(catalog_loc)
```

```python
ml.head()
```

```python

nsa = Table.read('/media/walml/beta1/galaxy_zoo/decals/catalogs/raw/nsa_v1_0_1.fits')

nsa = nsa[['RA', 'DEC', 'IAUNAME', 'PETRO_THETA', 'PETRO_TH50', 'PETRO_TH90', 'ELPETRO_ABSMAG', 'SERSIC_NMGY', 'Z', 'ELPETRO_MASS']]

#FNugriz

nsa['ELPETRO_ABSMAG_R'] = nsa['ELPETRO_ABSMAG'][:, 4]
nsa['SERSIC_NMGY_R'] = nsa['SERSIC_NMGY'][:, 4]
nsa['ELPETRO_ABSMAG_U'] = nsa['ELPETRO_ABSMAG'][:, 2]
nsa['SERSIC_NMGY_U'] = nsa['SERSIC_NMGY'][:, 2]
del nsa['ELPETRO_ABSMAG']
del nsa['SERSIC_NMGY']

nsa['REDSHIFT'] = nsa['Z']
del nsa['Z']

nsa = nsa.to_pandas()
for col in nsa.columns.values:
    nsa[col.lower()] = nsa[col]
    del nsa[col]

nsa['iauname'] = nsa['iauname'].str.decode('utf8')
assert not any(nsa['iauname'].duplicated())
```

```python
nsa["absmag_r"]=nsa["elpetro_absmag_r"]+5*np.log10(0.7)
nsa["absmag_u"]=nsa["elpetro_absmag_u"]+5*np.log10(0.7)
```

```python
nsa.columns.values
```

```python
ml.columns.values
```

```python
df = pd.merge(ml, nsa, on='iauname', how='inner')
df.columns.values
```

```python
df['u_minus_r'] = df['absmag_u'] - df['absmag_r']
```

```python
df = df.replace([np.inf, -np.inf], np.nan)
df = df[~pd.isna(df['absmag_r'])].reset_index()
```

```python
df['absmag_r'].hist(bins=30)
```

```python
df_sample = df.sample(50000).query('absmag_r < -21')
```

```python
def plot_valley(sample):
    fig, ax = plt.subplots()
    sns.kdeplot(np.log10(sample['elpetro_mass']), sample['u_minus_r'], ax=ax)
    ax.set_xlim([9.5, 11])
    ax.set_ylim([1., 3.])
    return ax
```

```python
_ = plot_valley(df_sample)
```

```python
cols_to_use = ['petro_theta', 'petro_th50', 'petro_th90', 'log_elpetro_mass', 'redshift']
# cols_to_use += [x for x in df.columns if '_fraction' in x]
cols_to_use += [
    'disk-edge-on_yes_fraction',
    'smooth-or-featured_smooth_fraction',
    'smooth-or-featured_artifact_fraction',
    'has-spiral-arms_yes_fraction',
    'bar_no_fraction',
    'bulge-size_moderate_fraction',
    'bulge-size_large_fraction',
    'edge-on-bulge_rounded_fraction',
#     'spiral-winding_medium_fraction',
    'merging_none_fraction'
]
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_sample[cols_to_use], how_green, test_size=0.33, random_state=42)
```
