### What do the galaxies look like?

```python
# temporarily pasted from illustris_to_tfrecord

def get_image_mask(img):
    return make_source_mask(img, snr=2, npixels=5, dilate_size=11)


def mask_background(img):
    return img * get_image_mask(img)


def clip_img(img):
    source_img = mask_background(img)  # doesn't mask, but uses mask to estimate source statistics
    return np.clip(img, a_min=0, a_max=source_img.mean() + source_img.std() * 8)


def normalise(img):
    return matplotlib.colors.Normalize(vmin=None, vmax=None, clip=False)(img)


def preprocess_img(img):
    return normalise(np.sqrt(clip_img(img)))


def load_illustris_as_pil(subject):  # should be dict with 'fits_loc' component
    img = fits.getdata(subject['fits_loc'])
    return Image.fromarray(preprocess_img(img) * 255)
```

```python
def show_examples(df, n_samples=10):
    fig, axes = plt.subplots(nrows=1, ncols=n_samples, figsize=(30, 8))
    subjects = df.sample(n_samples)
    for n in range(n_samples):
        subject = subjects.iloc[n]
        axes[n].imshow(load_illustris_as_pil(subject))
        axes[n].set_xlabel('Time: {}. Ratio: {}'.format(
            subject['max_time'],
            subject['mass_ratio']
        ))
    fig.tight_layout()
```

### What distributions of mass, mass ratio and max time does each category have?

```python
major = raw_metadata_with_names[raw_metadata_with_names['name'] == 'major_merger']
minor = raw_metadata_with_names[raw_metadata_with_names['name'] == 'minor_merger']
non = raw_metadata_with_names[raw_metadata_with_names['name'] == 'non_merger']
random = raw_metadata_with_names[raw_metadata_with_names['name'] == 'random_sample']
```

```python
categories = [major, minor, non, random]
```

```python
def show_hists(df_list, var_col, upper_lim=None):
    fig, axes = plt.subplots(ncols=len(df_list), figsize=(20, 4))
    for n, df in enumerate(df_list):
        data = df[var_col]
        if upper_lim:
            data = data[data < upper_lim]
        data.plot(kind='hist', ax=axes[n])
        axes[n].set_title(df.iloc[0]['name'])
        axes[n].set_xlabel(var_col)
    fig.tight_layout()
```

```python
show_hists(categories, 'mass', upper_lim=25)
```

All categories have a vaguely similar mass range on log scale. Major and Non have quite odd mass distributions.

```python
show_hists(categories, 'mass_ratio')
```

```python
non[non['mass_ratio'] < 0.02]['mass_ratio'].hist()
```

```python
random[random['mass_ratio'] < 0.02]['mass_ratio'].hist()
```

As hoped, the mass ratio scale is major > minor ( < 0.1 ) > non and random

```python
show_hists(categories, 'max_time')
```
